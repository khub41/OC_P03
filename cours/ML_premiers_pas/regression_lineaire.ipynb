```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('house.csv')
```

```python
data.shape
```

```python
plt.figure(figsize = (14,8))
plt.style.use('seaborn')
plt.scatter(data['surface'], data['loyer'], label='Loyers', alpha=0.7)
plt.xlabel("Surface du bien")
plt.ylabel("Valuer du loyer")
plt.legend()
plt.show()
```

```python
data = data[data['loyer'] <= 10000]
```

Entrainement

```python
X = np.matrix([np.ones(data.shape[0]), data['surface']]).T
y = np.matrix(data['loyer']).T
```

```python
X[:10]
```

```python
y[:10]
```

```python
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
```

```python
theta
```

```python
plt.figure(figsize = (8,6))
plt.style.use('seaborn')
plt.scatter(data['surface'], data['loyer'], label='Observ', alpha=0.7)
plt.plot([0,250], [theta.item(0), theta.item(0) + 250 * theta.item(1)], linestyle='--', color='black', label='Modle')
plt.xlabel("Surface du bien")
plt.ylabel("Valuer du loyer")
plt.legend()
plt.show()
```
